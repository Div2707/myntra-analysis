```python
#top 5 seller showing them through barplot
OG_sellers=myntra['seller'].value_counts().head(5)
plt.figure(figsize=(10,6))
sns.barplot(x=OG_sellers.index ,y=OG_sellers.values)
plt.xlabel('Seller')
plt.ylabel('Count')
plt.title('Top 5 Sellers by Value Counts')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
#bottom 5 seller showing them through barplot
noob_sellers=myntra['seller'].value_counts().tail(5)
plt.figure(figsize=(10,6))
sns.barplot(x=noob_sellers.index ,y=noob_sellers.values)
plt.xlabel('Seller')
plt.ylabel('Count')
plt.title('Top 5 Sellers by Value Counts')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
